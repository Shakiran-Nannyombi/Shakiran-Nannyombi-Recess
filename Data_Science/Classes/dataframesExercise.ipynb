```python
import pandas as pd

# read the CSV file into a DataFrame
df = pd.read_csv('files/AfricaCupofNationsMatches.csv')
```

```python
# Display the first 7 rows of the DataFrame
print(df.head(7))
```

```python
# print the selected names of the columns in the DataFrame
selected_columns = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
print(selected_columns.head())
```

```python
# display colunms with 'egypt' entry in either HomeTeam or AwayTeam
egypt_matches = df[(df['HomeTeam'] == 'Egypt') | (df['AwayTeam'] == 'Egypt')]
print(egypt_matches)
```

```python
# diplaying total number of rows and columns in the DataFrame
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
# Finding and display the rows where the 'Attendance' column is null
missing_attendance = df[df['Attendance'].isnull()]
print(missing_attendance)
```

```python
# displaying the rows where the 'HomeTeamGoals' is between 3 and 6
home_goals_between_3_and_6 = df[(df['HomeTeamGoals'] >= 3) & (df['HomeTeamGoals'] <= 6)]
print(home_goals_between_3_and_6)
```

```python
# changing the value of 'AwayTeamGoals' for the row with index 2
df.at[2, 'AwayTeamGoals'] = 10
print(df.iloc[2])
```

```python
# sorting the DataFrame by 'HomeTeam' and 'HomeTeamGoals' in descending order
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
print(sorted_df)
```

```python
# list all the column names in the DataFrame
print(df.columns.tolist())
```

```python
# Adding a new column 'GoalDifference' that calculates the difference between 'HomeTeamGoals' and 'AwayTeamGoals'
df['GoalDifference'] = df['HomeTeamGoals'] - df['AwayTeamGoals']
print(df.head())
```

```python
# adding two new rows to the DataFrame
new_rows = pd.DataFrame({
    'HomeTeam': ['TeamA', 'TeamB'],
    'AwayTeam': ['TeamC', 'TeamD'],
    'HomeTeamGoals': [1, 2],
    'AwayTeamGoals': [0, 1],
    'Attendance': [10000, 20000]
})
df = pd.concat([df, new_rows], ignore_index=True)
print(df.tail())
```

```python
# Changing the 'AwayTeam' name from 'Uganda' to 'China'
df.loc[df['AwayTeam'] == 'Uganda', 'AwayTeam'] = 'China'
print(df['AwayTeam'].value_counts())
```

```python
# resetting the index of the DataFrame
df.reset_index(drop=True, inplace=True)
print(df.head())
```

```python
# checking if the 'Stadium' column is present in the DataFrame
if 'Stadium' in df.columns:
    print("Stadium column is present")
else:
    print("Stadium column is not present")
```

```python
# converting the 'AwayTeamGoals' column to float type
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)
print(df['AwayTeamGoals'].dtype)
```

```python
# removing the last 10 rows from the DataFrame
df = df[:-10]
print(df.tail())
```

```python
# iterating through the DataFrame and printing the index, HomeTeam, and AwayTeam
for index, row in df.iterrows():
    print(f"Index: {index}, Home Team: {row['HomeTeam']}, Away Team: {row['AwayTeam']}")
```

```python
# changing order of columns in the DataFrame
new_ordered_columns = df[['AwayTeam', 'HomeTeam', 'AwayTeamGoals', 'HomeTeamGoals', 'Attendance']]
print(new_ordered_columns.head())
```

```python
#deleting DataFrame row(s) whose value is 0 in the ‘HomeTeamGoals’ columns.
df = df[df['HomeTeamGoals'] != 0]
print(df.head())
```
