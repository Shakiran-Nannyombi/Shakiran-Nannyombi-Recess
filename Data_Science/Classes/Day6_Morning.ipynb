```python
# Data Cleaning and Data Preprocessing
# Empty values
# Check for null values
df = pd.read_csv('files/data.csv')
print(df.isnull().sum())
```

```python
# check wrong data types
print(df.dtypes)
```

```python
# Data Cleaning and Data Preprocessing
# Empty values
# Check for null values
print(df.isnull().sum())
```

```python
# Fill empty values with mean for numeric columns
for col in ['Calories']:
    df[col].fillna(df[col].mode(), inplace=True)
```

```python
# Data Visualization and 

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#histograms

#set up the figure size
plt.figure(figsize=(12, 8))

# Duration histogram
plt.subplot(2, 2, 1)
sns.histplot(df['Duration'], bins=30, kde=True)
plt.title('Distribution of Duration')

# Calories histogram    
plt.subplot(2, 2, 2)
sns.histplot(df['Calories'], bins=30, kde=True)
plt.title('Distribution of Calories')

# Pulse histogram
plt.subplot(2, 2, 3)
sns.histplot(df['Pulse'], bins=30, kde=True)
plt.title('Distribution of Pulse')

# Maxpulse histogram    
plt.subplot(2, 2, 4)
sns.histplot(df['Maxpulse'], bins=30, kde=True)
plt.title('Distribution of Maxpulse')

plt.tight_layout()
plt.show()
```

```python
# set up the figure size for scatter plots
plt.figure(figsize=(15, 10))

# Scatter plot of Duration vs Pulse
plt.subplot(2, 2, 1)
sns.scatterplot(x='Duration', y='Pulse', data=df)
plt.title('Scatter Plot of Duration vs Pulse')

# Scatter plot of pulse vs Calories
plt.subplot(2, 2, 2)
sns.scatterplot(x='Pulse', y='Calories', data=df)
plt.title('Scatter Plot of Pulse vs Calories')

# Scatter plot of Maxpulse vs Calories
plt.subplot(2, 2, 3)
sns.scatterplot(x='Maxpulse', y='Calories', data=df)
plt.title('Scatter Plot of Maxpulse vs Calories')

# Scatter plot of Duration vs Calories
plt.subplot(2, 2, 4)
sns.scatterplot(x='Duration', y='Calories', data=df)
plt.title('Scatter Plot of Duration vs Calories')

plt.tight_layout()
plt.show()
```

```python
# set up the figure size for scatter plots
plt.figure(figsize=(12, 8))

# Scatter plot of Duration vs Maxpulse
plt.subplot(1, 2, 1)
sns.scatterplot(x='Duration', y='Maxpulse', data=df)    
plt.title('Scatter Plot of Duration vs Maxpulse')

# Scatter plot of Pulse vs Maxpulse
plt.subplot(1, 2, 2)
sns.scatterplot(x='Pulse', y='Maxpulse', data=df)
plt.title('Scatter Plot of Pulse vs Maxpulse')
```

```python
# correlation heatmap
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
